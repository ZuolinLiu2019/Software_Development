### Introduction

Data Source: https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data

This dataset contains 79 explanatory variables describing (almost) every aspect of residential homes in Ames, Iowa. The goal is to predict the house sale price.

In this notebook, we will proceed to build a linear regression model to predict the house price using the feature selected from the previous section.

```python
# Import libraries
import pandas as pd
import numpy as np

# for plotting
import matplotlib.pyplot as plt
%matplotlib inline

# to build the models
from sklearn.linear_model import Lasso

# to evaluate the models
from sklearn.metrics import mean_squared_error
from math import sqrt

# to visualise al the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
# load dataset prepared by the feature engineering section

X_train = pd.read_csv('xtrain.csv')
X_test = pd.read_csv('xtest.csv')

X_train.head()
```

```python
# keep the target separated
y_train = X_train['SalePrice']
y_test = X_test['SalePrice']
```

```python
# load selected features from the feature selection section

features = pd.read_csv('selected_features.csv', header=None)
features = [x for x in features[0]] 

# I will add 'LotFrontage' as a predictor because it seems pretty relevant

features = features + ['LotFrontage'] 

features
```

```python
# extract columns using the selected feature for both trainning and testing dataset

X_train = X_train[features]
X_test = X_test[features]
```

### Regularized linear regression

```python
# train the model
lin_model = Lasso(alpha=0.005, random_state=0) # set the random_state / seed
lin_model.fit(X_train, y_train)
```

```python
# Model evaluation:
# since the target was transfromed into log scale, we need to transform the predictions back to the original house
# price value to calaculate the metrics (mean sqaured error and rooted mean square error)

pred = lin_model.predict(X_train)
print(f'Trainning mse: {mean_squared_error(np.exp(y_train), np.exp(pred))}')
print(f'Trainning rmse: {sqrt(mean_squared_error(np.exp(y_train), np.exp(pred)))}\n')
pred = lin_model.predict(X_test)
print(f'Testing mse: {mean_squared_error(np.exp(y_test), np.exp(pred))}')
print(f'Testing rmse: {sqrt(mean_squared_error(np.exp(y_test), np.exp(pred)))}\n')
print('Average house price: ', np.exp(y_train).median())
```

```python
# let's evaluate our predictions respect to the original price
# they should aligned on the 1:1 line
plt.scatter(y_test, lin_model.predict(X_test))
plt.plot(np.linspace(11.25, 13, 10), np.linspace(11.25, 13, 10), 'k-')
plt.xlabel('True House Price')
plt.ylabel('Predicted House Price')
plt.legend(['1:1'])
plt.title('Evaluation of Lasso Predictions')
```

```python
# let's evaluate the distribution of the errors: 
# they should be fairly normally distributed

errors = y_test - lin_model.predict(X_test)
errors.hist(bins=15)
```

```python
# Finally, just for fun, let's look at the feature importance

importance = pd.Series(np.abs(lin_model.coef_.ravel()))
importance.index = features
importance.sort_values(inplace=True, ascending=False)
importance.plot.bar(figsize=(18,6))
plt.ylabel('Lasso Coefficients')
plt.title('Feature Importance')
```
